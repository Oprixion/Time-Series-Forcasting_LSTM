```python
import pandas as pd
```

# EDA

```python
path = "Data\\energy_weather_featured.csv"
df = pd.read_csv(path)
df.columns
```

```python
# Drop additional redundant columns due to serving its purpose in the feature engineering process
# Keepig Date_Begin_Local to use as a reference, DROP BEFORE TRAINING!!!

columns_to_drop = ["ACTUAL_POOL_PRICE", "ACTUAL_AIL", "T2M_CALGARY", "WS50M_CALGARY", 'T2M_CALGARY', 'WS50M_CALGARY', 'T2M_LETHBRIDGE', 'WS50M_LETHBRIDGE']
df.drop(columns=columns_to_drop, inplace=True)
print(df.head())
print(df.columns)
print(df.shape)
```

```python
# Train-test split (chronological — no shuffling)
split_idx = int(len(df) * 0.8)
train_df = df.iloc[:split_idx]
test_df  = df.iloc[split_idx:]
print(f"Train shape: {train_df.shape}")
print(f"Test shape:  {test_df.shape}")
print(f"Train period: {train_df['Date_Begin_Local'].iloc[0]} → {train_df['Date_Begin_Local'].iloc[-1]}")
print(f"Test period:  {test_df['Date_Begin_Local'].iloc[0]} → {test_df['Date_Begin_Local'].iloc[-1]}")
```

```python
# Export train and test splits
train_df.to_csv("Data\\train.csv", index=False)
test_df.to_csv("Data\\test.csv", index=False)
print("Saved Data\\train.csv and Data\\test.csv")
```
